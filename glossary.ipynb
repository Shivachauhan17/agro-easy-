```python
import nltk
```

```python
import re
```

```python
glossary=nltk.sent_tokenize(str(glossary))
glossary
```

```python
for i in range(0,len(glossary)):
    glossary[i]=glossary[i].replace('\n',' ')
```

```python
glossary
```

```python
all=[]
for i in range(0,len(glossary)):
    all.append(re.sub(r'^[^:]*$', '',glossary[i]))
all
```

```python
len(all)
```

```python
non_empty_all=[]
for i in all:
    if i!='':
        non_empty_all.append(i)
len(non_empty_all)
```

```python
non_empty_all
```

```python
words=[]
for i in range(0,len(non_empty_all)):
    word=re.sub(r':.*', '', non_empty_all[i])
    words.append(word)
words
```

```python
words='\n'.join(words)
words    
```
